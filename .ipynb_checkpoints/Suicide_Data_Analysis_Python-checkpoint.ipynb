```python
D = data.groupby("Year")[["Total"]].sum()
```

```python
x = D.index.values.reshape(-1,1)
y = D.values
```

```python
plt.scatter(x,y)
plt.xlabel('Year')
plt.ylabel('Total suicide')
plt.show()
```

```python


counts = data.groupby("Gender").Total.sum()
print(counts)
# Plot a pie chart
counts.plot(kind='pie', title='Gender Count',figsize=(10,8))

plt.legend()
plt.show()

```

```python
cause = data[data['Type_code'] == 'Causes']
edu_status = data[data['Type_code'] == 'Education_Status']
means_adpt = data[data['Type_code'] == 'Means_adopted']
prof = data[data['Type_code'] == 'Professional_Profile']
soc_status = data[data['Type_code'] == 'Social_Status']
```

```python
def plot_type(data, Title, X_lab):
    p_type = data.groupby('Type').sum()['Total']
    sort_df = p_type.sort_values(ascending = False)

    fig = sort_df.plot(kind='bar', figsize = (10,6), title = Title + '\n', width = 0.75)
    fig.set_xlabel('\n' + X_lab )
    fig.set_ylabel('Count\n')
    sns.set_style('whitegrid')
    sns.set_palette('Set2')  
```

```python
plot_type(cause, 'Suicide by cause', 'Cause')
```

```python
plot_type(edu_status, 'Suicide by Education Status', 'Education Status')

```

```python
plot_type(means_adpt, 'Suicide by Means Adopted', 'Means Adopted')
```

```python
plot_type(prof, 'Suicide by Professional Profile', 'Professional Profile')
```

```python
plot_type(soc_status, 'Suicide by Social Status', 'Social Status')
```

```python
#CALLING LINAEAR REGRESSION
#BEST FIT LINE
#
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_#slope of best fit line
```

```python
reg.intercept_#intercept value i.e value of c in y=mx+c
```

```python
y
```

```python
#PREDICTING VALUE OF Y BY BEST FIT LINE
```

```python
Yp = reg.predict(x)
```

```python
#SHOWING THE BEST FIT LINE
```

```python
plt.scatter(x,y)
plt.plot(x,Yp)
plt.show()
```

```python
reg.score(x,y)*100    #ACCURACY OF THIS MODEL IN PERCENTAGE
```

```python
reg.predict([[2021]]) ####PREDICTING VALUE FOR 2021
```

```python
from sklearn.preprocessing import PolynomialFeatures ##import polynomial
poly = PolynomialFeatures(degree = 5) #storing poly features
X = poly.fit_transform(x)
pd.DataFrame(X)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
y
```

```python
x
```

```python
YP = reg.predict(X)
```

```python
plt.scatter(x,y)
plt.plot(x,YP)
plt.show()
```

```python
reg.score(X,y)*100
```

```python
value = poly.transform([[2021]])
```
