```python
col=['cartodb_id','cri_score','fatalities_per_100k_total','fatalities_total','losses_per_gdp__total','losses_usdm_ppp_total','rw_country_code']

df= real_data[col]
```

```python
from sdv.metadata import SingleTableMetadata #this is a function that creates metadata, tries to capture the datatypes

metadata = SingleTableMetadata()#creating an empty data table
metadata.detect_from_csv(filepath='climate-risk-index-1.csv')
```

```python
metadata
```

```python
#editing the metadata file
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata.load_from_json(
    filepath='crel_m.json')
```

```python
metadata.validate()
```

```python
#applying CTGAN algorithm to generate synthetic data


from sdv.single_table import CTGANSynthesizer

synthesizer = CTGANSynthesizer(
    metadata, # required
    enforce_rounding=False,
    epochs=1000,
    verbose=True
)
```

```python
##CTGAN

synthesizer.fit(df)
```

```python
sdf=synthesizer.sample(num_rows=1000)
```

```python
sdf
```

```python
df.isnull().sum()
```

```python
sdf.isnull().sum()
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(
    real_data=df,
    synthetic_data=sdf,
    metadata=metadata)
```

```python
##GAUSSIAN COPULA

from sdv.single_table import GaussianCopulaSynthesizer

synthesizer2 = GaussianCopulaSynthesizer(metadata)
synthesizer2.fit(df)

sdf2 = synthesizer2.sample(num_rows=1000)
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report2 = evaluate_quality(
    real_data=df,
    synthetic_data=sdf2,
    metadata=metadata)
```

```python
#missing value imputation by conditional sampling(CTGAN METHOD)

reference_data = pd.DataFrame(data={
    'cri_score': [124.5, 109.17, 117],
    'fatalities_per_100k_total': [0.00, 0.03, 0.01],
    'fatalities_total' : [0, 3, 2],
    'losses_usdm_ppp_total' : [0, 0.087, 0.276]
})

sdf_miss = synthesizer.sample_remaining_columns(
    known_columns=reference_data,
    max_tries_per_batch=1000,
    batch_size=1000
)

```

```python
sdf_miss
```

```python
#missing value imputation by Gaussian Copula method

reference_data = pd.DataFrame(data={
    'cri_score': [124.5, 109.17, 117],
    'fatalities_per_100k_total': [0.00, 0.03, 0.01],
    'fatalities_total' : [0, 3, 2],
    'losses_usdm_ppp_total' : [0, 0.087, 0.276]
})

sdf_miss2 = synthesizer2.sample_remaining_columns(
    known_columns=reference_data,
    max_tries_per_batch=1000,
    batch_size=1000
)
```

```python
sdf_miss2
```

```python
##CONDITIONAL SAMPLE(SCENARIO)

from sdv.sampling import Condition

zero_fatalitites_losses_usdm = Condition(
    num_rows=100,
    column_values={'fatalities_total': 0, 'losses_usdm_ppp_total': 0}
)

non_zero = Condition(
    num_rows=10,
    column_values={'fatalities_total': 3}
)
synthetic_data = synthesizer2.sample_from_conditions(
    conditions=[zero_fatalitites_losses_usdm,non_zero],

)


```
