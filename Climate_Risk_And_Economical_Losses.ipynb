```python
real_data=pd.read_csv('climate-risk-index-1.csv')
real_data.head()
```

```python
real_data=real_data.drop(['the_geom','the_geom_webmercator'],axis=1)
real_data
```

Dropping columns that are corresponding to rank columns as rank are determined on yhe basis of corresponding score columns.

```python
col=['cartodb_id','cri_score','fatalities_per_100k_total','fatalities_total','losses_per_gdp__total','losses_usdm_ppp_total','rw_country_code']

df= real_data[col]
```

```python
from sdv.metadata import SingleTableMetadata #this is a function that creates metadata, tries to capture the datatypes

metadata = SingleTableMetadata()#creating an empty data table
metadata.detect_from_csv(filepath='climate-risk-index-1.csv')
```

```python
metadata
```

```python
#editing the metadata file
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata.load_from_json(
    filepath='crel_m.json')
```

```python
metadata.validate()
```

```python
#applying CTGAN algorithm to generate synthetic data


from sdv.single_table import CTGANSynthesizer

synthesizer = CTGANSynthesizer(
    metadata, # required
    enforce_rounding=False,
    epochs=1000,
    verbose=True
)
```

```python
##CTGAN

synthesizer.fit(df)
```

```python
sdf=synthesizer.sample(num_rows=1000)
```

```python
sdf
```

```python
df.isnull().sum()
```

```python
sdf.isnull().sum()
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(
    real_data=df,
    synthetic_data=sdf,
    metadata=metadata)
```

```python
##GAUSSIAN COPULA

from sdv.single_table import GaussianCopulaSynthesizer

synthesizer2 = GaussianCopulaSynthesizer(metadata)
synthesizer2.fit(df)

sdf2 = synthesizer2.sample(num_rows=1000)
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report2 = evaluate_quality(
    real_data=df,
    synthetic_data=sdf2,
    metadata=metadata)
```
